# NBA Chat Assistant

An intelligent chatbot that answers questions about NBA using:
- **RAG (Retrieval Augmented Generation)** for accurate information
- **LangGraph** for conversation management
- **LangChain** for LLM integration

## Features

- Answers questions about NBA players, teams, stats, history
- Retrieves information from trusted NBA sources
- Maintains conversation context
- Provides source citations
- Can integrate with NBA APIs for real-time data
